Unlike the name suggest, Logistic regression is a popular binary classification algorithm that models the probability of the occurrence of a binary event. The logistic function, also known as the sigmoid function, is commonly used in logistic regression.

```python
import numpy as np
```

The sigmoid function is used to squash the output of the linear combination of input features and weights to a range between 0 and 1. It's a key component in logistic regression as it transforms any real-valued number into a probability.

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

This function initializes the weights and bias to zero. These are the parameters that the algorithm will learn during the training process.

```python
def hypothesis(num_features):
    weights = np.zeros(num_features)
    bias = 0
    return weights, bias
```

This function calculates the gradients of the loss with respect to the weights and bias. These gradients are then used to update the parameters during the optimization process.

```python
def gradients(X, y, weights, bias):
    num_samples = X.shape[0]

    # Compute predicted probabilities
    y_pred = sigmoid(np.dot(X, weights) + bias)

    # Compute gradients
    dw = (1/num_samples) * np.dot(X.T, (y_pred - y))
    db = (1/num_samples) * np.sum(y_pred - y)

    return dw, db
```

This function adjusts the weights and bias based on the computed gradients and a specified learning rate. This step is part of the gradient descent optimization process.

```python
def gradient_descent(weights, bias, dw, db, learning_rate):
    weights -= learning_rate * dw
    bias -= learning_rate * db

    return weights, bias
```

The logistic_regression function puts everything together. It initializes parameters, performs gradient descent for a specified number of iterations, and returns the learned weights and bias.

```python
def logistic_regression(X, y, learning_rate=0.01, num_iterations=1000):
    num_samples, num_features = X.shape

    # Initialize parameters
    weights, bias = hypothesis(num_features)

    # Gradient Descent
    for _ in range(num_iterations):
        # Compute gradients
        dw, db = gradients(X, y, weights, bias)

        # Update parameters
        weights, bias = gradient_descent(weights, bias, dw, db, learning_rate)

    return weights, bias
```

The predict function takes the learned weights and bias and uses them to make predictions on new data. It converts the predicted probabilities into binary predictions by applying a threshold (0.5 in this case).

```python
def predict(X, weights, bias):
    # Predict probabilities
    y_pred = sigmoid(np.dot(X, weights) + bias)

    # Convert probabilities to binary predictions (0 or 1)
    predictions = [1 if i > 0.5 else 0 for i in y_pred]

    return predictions
```
